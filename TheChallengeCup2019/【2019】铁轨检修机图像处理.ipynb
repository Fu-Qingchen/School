```
import math
import numpy as np
import matplotlib.pyplot as plt
import urllib
import cv2
from ipywidgets import interact
import ipywidgets as widgets
```

## 获取图像

```
origin_img = cv2.imread("Demo.jpg")
plt.imshow(origin_img, cmap=plt.cm.binary)
plt.grid(False)
```

## 图像增强

```
enhancement_img = origin_img.copy()
rows,cols,channels=origin_img.shape
a=3.5
b=50
for i in range(rows):
    for j in range(cols):
        for c in range(1):
            color=origin_img[i,j][c]*a+b
            if color>255:
                enhancement_img[i,j][c]=255
            elif color<0:
                enhancement_img[i,j][c]=0
plt.imshow(enhancement_img)
plt.grid(False)
```

## 轨面分割

```
def f1(x):
  threshold_image = cv2.cvtColor(enhancement_img, cv2.COLOR_BGR2GRAY)
  #阈值化处理，阈值为：105
  (T, thresh) = cv2.threshold(threshold_image, x, 255, cv2.THRESH_BINARY)
  plt.imshow(thresh)
  plt.grid(False)
interact(f1,x = 105)
```

```
sum_list = thresh.sum(axis = 1)
plt.plot(sum_list)
```

```
cropped_img = origin_img[80:160, :]
plt.imshow(cropped_img)
plt.grid(False)
```

## 边缘检测

```
canny_img = cv2.Canny(cropped_img, 70, 200)
plt.imshow(canny_img)
plt.grid(False)
```
